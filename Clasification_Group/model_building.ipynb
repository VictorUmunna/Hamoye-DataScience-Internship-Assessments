```python
# Getting model log summaries with major focus on coefficients
import statsmodels.api as sm

# building the model and fitting the data
log_reg = sm.Logit(y2, x2).fit()

#printing the log summary
print(log_reg.summary())

# Determining Model Accuracy
# fit a logistic regression on the training dataset
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
# fit model
model.fit(x2, y2)
# make predictions
yhat = model.predict(x2)
# evaluate predictions
acc = accuracy_score(y2, yhat)
print(acc)
```

```python
# User Prompt

# Redefining the dataset by only retaining valid columns
import csv

_questions = ['Q42', 'Q48', 'Q53', 'Q47', 'Q52', 'Q45', 'Q43', 'Q7', 'Q46', 'Q6']

print('Please answer the following questions truthfully.')
print('All responses are collected on a 5 point scale \n 0=Never, 1=Seldom, 2=Averagely, 3=Frequently, 4=Always\n')

# user input prompt function
def prompt_for_answers():
    _responses = []
    with open("reference.tsv", "r") as f:
        data = csv.DictReader(f, delimiter='|', quoting=csv.QUOTE_NONE)
        for i in data:
            if "Q{}".format(i['atribute_id']) in _questions:
                answer = input(i['description'] + " ")
                if answer: _responses.append(answer)
    return _responses
                
answer_list = prompt_for_answers()

print(answer_list)
```

```python
# Getting predictions on user input
# Defining user input
lst = answer_list

# converting user input list to array
arr = np.array(lst)

# Reshaping the user input array from 1D to 2D 
newarr = arr.reshape(1, -1)

#getting likelihood of divorce 
result = model.predict(newarr)

# values between 0 and 0.5 will be considered 0 (Not Divorced)
# values between 0.5 and 1, will be considered 1 (Divorced)
if result < 0.5:
  print(result, 'Not likely to Divorce.')
  print('Model accuracy = ', acc, '\n')
else:
  print(result, 'Likely to Divorce')
  print('Model accuracy = ', acc, '\n')
```
